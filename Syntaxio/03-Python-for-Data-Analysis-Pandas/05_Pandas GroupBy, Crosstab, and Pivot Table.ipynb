# Groupby

The groupby method allows you to group rows of data together and call aggregation functions


```python
import pandas as pd
import numpy as np


pd.set_option(
    'display.width', 10000
)
pd.set_option(
    'display.max_columns', 50
)
pd.set_option(
    'styler.render.max_columns', 50
)
pd.set_option(
    'styler.render.max_elements', 10000
)

pd.set_option(
    'display.max_rows', 100
)

# vérifier la version
pd.__version__
```

```python
df = pd.read_csv("data/heart.csv")
df
```

## groupby object

```python
g = df.groupby('sexe')
```

```python
type(g)
```

Select one of the group

```python
df_for_group = g.get_group('masculin')
df_for_group
```

```python
type(g.get_group('masculin'))
```

```python
g.get_group('masculin').shape
```

```python
g.get_group('masculin')['age'].mean()
```

```python
g.get_group("feminin")["age"].mean()
```

### Aggregation functions on GroupBy

```python
g.mean(numeric_only=True)
```

```python
g[["age", "depression"]].agg(
    [
        pd.Series.mean,
        pd.Series.std,
		pd.Series.min,
		pd.Series.max,
        pd.Series.sum,
        pd.Series.count,
    ]
)
```

```python
g.describe()
```

```python
g.describe().transpose()
```

```python
g.describe().transpose()['masculin']
```

### Iterate on groups

```python
for groupe in g:
    #groupe is a tuple
    print(f"{groupe[0] = }") # group label
    #access to the 'age' variable of the concerned group
    print(f"{pd.Series.mean(groupe[1]['age']) = }")
    print()
```

## Crosstab

A bit more general than groupby, it allows to compute a contingency table.

```python
pd.crosstab(df['sexe'],df['coeur'])
```

### Normalize

```python
pd.crosstab(df['sexe'],df['coeur'],normalize='index')
```

```python
pd.crosstab(df['sexe'],df['coeur'],normalize='columns')
```

```python
pd.crosstab(df['sexe'],df['coeur'],normalize='all')
```

### Aggregate function

```python
pd.crosstab(df['sexe'],df['coeur'],values=df['age'],aggfunc=pd.Series.median)
```

### More than 2 dimensions:

```python
pd.crosstab(
    [df['sexe'], df['sucre']],
    df['coeur'],
    normalize='index'
)
```

### Pivot Table

A pivot table is a table that summarizes data from a larger dataset. 

It is a way to reorganize and summarize data in a more intuitive and user-friendly format.


```python
data = {
	'A':['foo','foo','foo','bar','bar','bar'],
	'B':['one','one','two','two','one','one'],
	'C':['x','y','x','y','x','y'],
	'D':[1,3,2,5,4,1]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.pivot_table(values='D',index=['A', 'B'],columns=['C'])
```

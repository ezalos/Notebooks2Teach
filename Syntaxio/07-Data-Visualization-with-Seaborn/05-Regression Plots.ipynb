# Regression Plots

Seaborn has many built-in features for regression plots, but we won't really discuss regression until the machine learning section of the course, so we'll only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also allows you to split these plots based on features, as well as color the hue based on features.

Let's explore how this works:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

Plot the data and a linear model

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Working with markers

The lmplot kwargs are passed to **regplot**, which is a more general form of `lmplot()`. 

`regplot()` has a scatter_kws parameter that is passed to plt.scatter. 

So you want to set the s parameter in this dictionary, which (somewhat confusingly) corresponds to the marker size squared. 

In other words, you end up passing a dictionary with the basic matplotlib arguments, in this case, s for the size of a scatter plot. 

Generally, you probably won't remember this by heart, but rather refer to the documentation.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

## Utiliser une grille

Nous pouvons ajouter une séparation plus variable dans les colonnes et les lignes à l'aide d'une grille. Indiquez-le simplement avec les arguments col ou row :

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted using the **height** and **aspect** parameters:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', palette='coolwarm',
           aspect=0.6, height=8)
```
